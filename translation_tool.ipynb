```python
# Step 1: Install required packages (run this cell once)
!pip install googletrans==4.0.0-rc1
!pip install gtts

# Step 2: Import libraries
import tkinter as tk
from tkinter import ttk, messagebox
from googletrans import Translator
from gtts import gTTS
import os

# Step 3: Define translation and speech functions
def translate_text():
    text = text_input.get("1.0", tk.END).strip()
    dest_lang = lang_var.get()
    if not text:
        messagebox.showwarning("Warning", "Please enter text to translate")
        return
    try:
        translator = Translator()
        translated = translator.translate(text, dest=dest_lang)
        result_output.config(state=tk.NORMAL)
        result_output.delete("1.0", tk.END)
        result_output.insert(tk.END, translated.text)
        result_output.config(state=tk.DISABLED)
        
        # Convert to speech
        tts = gTTS(translated.text, lang=dest_lang)
        tts.save("translated.mp3")
        os.system("start translated.mp3")  # Windows: use "start", Mac: "afplay", Linux: "mpg321"
    except Exception as e:
        messagebox.showerror("Error", str(e))

# Step 4: Build GUI
root = tk.Tk()
root.title("Text Translator & Speech")
root.geometry("500x400")

# Input text
tk.Label(root, text="Enter text:").pack(pady=5)
text_input = tk.Text(root, height=5)
text_input.pack(pady=5)

# Language selection
tk.Label(root, text="Select language (e.g., 'hi' for Hindi, 'fr' for French):").pack(pady=5)
lang_var = tk.StringVar(value='hi')
lang_entry = ttk.Entry(root, textvariable=lang_var)
lang_entry.pack(pady=5)

# Translate button
translate_btn = ttk.Button(root, text="Translate & Speak", command=translate_text)
translate_btn.pack(pady=10)

# Output
tk.Label(root, text="Translated Text:").pack(pady=5)
result_output = tk.Text(root, height=5, state=tk.DISABLED)
result_output.pack(pady=5)

root.mainloop()
```
